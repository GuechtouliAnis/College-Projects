```python
# Importation des biblioth√®ques
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Creation de graphe G
# Les sommets
G = nx.Graph()
sommets = [
    "KOR", "JAP", "USA", "MEX", "CAN", "NZE", "AUS",
    "IRE", "UK", "BEL", "FRA", "POR", "SPA", "SWI",
    "ITA", "TUR", "GER", "AUT", "GRE", "POL"]
# AUS : Australia
# AUT : Austria
G.add_nodes_from(sommets)

# Les aretes
aretes_G = [
    ('KOR','JAP'),('KOR','USA'),
    ('JAP','USA'),
    ('MEX','USA'),
    ('USA','CAN'),('USA','UK'),('USA','ITA'),('USA','GER'),('USA','POL'),
    ('CAN','UK'),('CAN','ITA'),
    ('UK','IRE'),('UK','SPA'),('UK','FRA'),('UK','AUS'),
    ('AUS','NZE'),
    ('FRA','BEL'),('FRA','POR'),('FRA','ITA'),('FRA','SPA'),('FRA','TUR'),
    ('TUR','GER'),
    ('GER','SPA'),('GER','ITA'),('GER','GRE'),('GER','AUT'),('GER','POL'),
    ('ITA','AUS'),('ITA','SWI')]

G.add_edges_from(aretes_G)
```

```python
# 1. D√©terminer et afficher la matrice d‚Äôadjacence de G
adj_matrix_G = nx.adjacency_matrix(G).todense()

nodes_G = list(G.nodes())
df_adj_G = pd.DataFrame(adj_matrix_G, index=nodes_G, columns=nodes_G)

print(f"Matrice d'adjacence de G:")
df_adj_G
```

```python
# Affichage de graphe G
pos = nx.spring_layout(G, seed=21)
nx.draw(G,
        pos,
        with_labels=True,
        node_color='red',
        font_weight='bold',
        font_size=10,
        node_size=500)
plt.title("Figure 1")
plt.show()
```

```python
# 3. Affichage de coefficient de clustering de chaque noeud
clust_coeff = nx.clustering(G)
df_clust_coeff = pd.DataFrame(clust_coeff.items(), columns=['Noeud', 'Coefficient de clustering'])
print(f"Le coefficient de clustering de chaque noeud:")
df_clust_coeff
```

```python
# Coefficient de clustering sans la fonction predefinie
def clustering_coefficient(G, node):
    neighbors = list(G.neighbors(node))
    k = len(neighbors)
    if k < 2:
        return 0
    edges = 0
    for i in range(k):
        for j in range(i+1, k):
            if G.has_edge(neighbors[i], neighbors[j]):
                edges += 1
    cc = 2 * edges / (k * (k-1))
    return cc

clust_coeff = {node: clustering_coefficient(G, node) for node in G.nodes()}
df_clust_coeff = pd.DataFrame(clust_coeff.items(), columns=['Noeud', 'Coefficient de clustering'])
print(f"Le coefficient de clustering de chaque noeud sans la fonction predefinie:")
df_clust_coeff
```

```python
# 4. Coefficient de clustering de graphe G
clust_coeff_G = nx.average_clustering(G)
print(f"Le coefficient de clustering de G est: {clust_coeff_G: .4f}")
```

```python
# Coefficient de clustering de graphe G sans la fonction predefinie
def clustering_coefficient_G(G):
    nodes = list(G.nodes())
    sum_clust_coeff = 0
    for node in nodes:
        sum_clust_coeff += clustering_coefficient(G, node)
    return sum_clust_coeff / len(nodes)

clust_coeff_G = clustering_coefficient_G(G)
print(f"Le coefficient de clustering de G sans la fonction predefinie est: {clust_coeff_G: .4f}")
```

```python
# 5. Degr√© de centralit√© de chaque noeud avec la fonction predefinie
deg_cent = nx.degree_centrality(G)
df_deg_cent = pd.DataFrame(deg_cent.items(), columns=['Noeud', 'Degr√© de centralit√©'])
print(f"Le degr√© de centralit√© de chaque noeud:")
df_deg_cent
```

```python
# Degr√© de centralit√© de chaque noeud sans la fonction predefinie
def degree_centrality(G, node):
    return len(list(G.neighbors(node))) / (len(G.nodes()) - 1)

deg_cent = {node: degree_centrality(G, node) for node in G.nodes()}
df_deg_cent = pd.DataFrame(deg_cent.items(), columns=['Noeud', 'Degr√© de centralit√©'])
print(f"Le degr√© de centralit√© de chaque noeud sans la fonction predefinie:")
df_deg_cent
```

```python
# 6. Le noeud le plus important
most_important_node = max(deg_cent, key=deg_cent.get)
print(f"Le noeud le plus important est: {most_important_node}")
# Justification:
print('Justification: Le noeud le plus important est celui qui a le plus grand degr√© de centralit√©,')
print('c\'est-√†-dire le noeud qui est le plus connect√© aux autres noeuds.')
```

```python
# 7. Centralit√© interm√©diaire de chaque noeud avec la fonction predefinie
bet_cent = nx.betweenness_centrality(G)
df_bet_cent = pd.DataFrame(bet_cent.items(), columns=['Noeud', 'Centralit√© interm√©diaire'])
print(f"La centralit√© interm√©diaire de chaque noeud:")
df_bet_cent
```

```python
# Centralit√© interm√©diaire de chaque noeud sans la fonction predefinie
def betweenness_centrality(G, node):
    nodes = list(G.nodes())
    sum_shortest_paths = 0
    for i in range(len(nodes)):
        for j in range(i+1, len(nodes)):
            if nx.has_path(G, nodes[i], nodes[j]):
                shortest_path = nx.shortest_path(G, nodes[i], nodes[j])
                if node in shortest_path:
                    sum_shortest_paths += 1
    return sum_shortest_paths / ((len(nodes) - 1) * (len(nodes) - 2))

bet_cent = {node: betweenness_centrality(G, node) for node in G.nodes()}
df_bet_cent = pd.DataFrame(bet_cent.items(), columns=['Noeud', 'Centralit√© interm√©diaire'])
print(f"La centralit√© interm√©diaire de chaque noeud sans la fonction predefinie:")
df_bet_cent
```

```python
# 8. Le noeud le plus influent
most_influential_node = max(bet_cent, key=bet_cent.get)
print(f"Le noeud le plus influent est: {most_influential_node}")
# Justification:
print('Justification: Le noeud le plus influent est celui qui a la plus grande centralit√© interm√©diaire,')
print('c\'est-√†-dire le noeud qui est le plus souvent sur les plus courts chemins entre les autres noeuds.')
```

```python
# Creation d'un graphe orient√© A
A = nx.DiGraph()

# Les sommets
sommets_A = [1,2,3,4,5,6,7,8]
A.add_nodes_from(sommets_A)

# 1:A / 2:B / 3:C / 4:D / 5:E / 6:F / 7:G / 8:H
# Les aretes
aretes_A = [
    (1,2),(1,4),(1,6),(1,5),
    (2,3),(2,8),
    (3,1),(3,5),(3,8),
    (4,3),(4,5),
    (5,8),(5,7),(5,6),
    (6,4),
    (7,6),(7,4),(7,8),
    (8,7)]

A.add_edges_from(aretes_A)

# Visualisation de graphe A
pos = nx.spring_layout(A, seed=24)
nx.draw(A,
        pos,
        with_labels=True,
        node_color='lightgreen',
        font_weight='bold',
        font_size=10,
        node_size=500)
plt.title("Graphe A")
plt.show()
```

```python
# 2. Matrice d'adjacence de A
adj_matrix_A = nx.adjacency_matrix(A).todense()

nodes_A = list(A.nodes())
df_adj_A = pd.DataFrame(adj_matrix_A, index=nodes_A, columns=nodes_A)

print(f"Matrice d'adjacence de A:")
df_adj_A
```

```python
# 3. Matrice transpos√©e de A
A_transpose = A.reverse()
adj_matrix_A_transpose = nx.adjacency_matrix(A_transpose).todense()
adj_matrix_A_transpose = pd.DataFrame(adj_matrix_A_transpose, index=nodes_A, columns=nodes_A)
adj_matrix_A_transpose
```
