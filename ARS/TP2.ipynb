```python
# Creation de graphe G
# Les sommets
G = nx.Graph()
sommets = [
    "KOR", "JAP", "USA", "MEX", "CAN", "NZE", "AUS",
    "IRE", "UK", "BEL", "FRA", "POR", "SPA", "SWI",
    "ITA", "TUR", "GER", "AUT", "GRE", "POL"]
# AUS : Australia
# AUT : Austria
G.add_nodes_from(sommets)

# Les aretes
aretes_G = [
    ('KOR','JAP'),('KOR','USA'),
    ('JAP','USA'),
    ('MEX','USA'),
    ('USA','CAN'),('USA','UK'),('USA','ITA'),('USA','GER'),('USA','POL'),
    ('CAN','UK'),('CAN','ITA'),
    ('UK','IRE'),('UK','SPA'),('UK','FRA'),('UK','AUS'),
    ('AUS','NZE'),
    ('FRA','BEL'),('FRA','POR'),('FRA','ITA'),('FRA','SPA'),('FRA','TUR'),
    ('TUR','GER'),
    ('GER','SPA'),('GER','ITA'),('GER','GRE'),('GER','AUT'),('GER','POL'),
    ('ITA','AUS'),('ITA','SWI')]

G.add_edges_from(aretes_G)
```

```python
# 1. Déterminer et afficher la matrice d’adjacence de G
adj_matrix_G = nx.adjacency_matrix(G).todense()

nodes_G = list(G.nodes())
df_adj_G = pd.DataFrame(adj_matrix_G, 
                        index=nodes_G,
                        columns=nodes_G)

print(f"Matrice d'adjacence de G:")
df_adj_G
```

```python
# Affichage de graphe G
pos = nx.spring_layout(G, seed=21)
nx.draw(G,
        pos,
        with_labels=True,
        node_color='red',
        font_weight='bold',
        font_size=10,
        node_size=500)
plt.title("Figure 1")
plt.show()
```

```python
# 3. Affichage de coefficient de clustering de chaque noeud
# Avec la fonction predefinie
cc = nx.clustering(G)
n_cc = pd.DataFrame(cc.items(),
                    columns=['Noeud', 'Coefficient de clustering'])
print(f"Le coefficient de clustering de chaque noeud:")
n_cc
```

```python
# Coefficient de clustering sans la fonction predefinie
def clustering_coefficient(G, node):
    neighbors = list(G.neighbors(node))
    k = len(neighbors)
    if k < 2:
        return 0
    edges = 0
    for i in range(k):
        for j in range(i+1, k):
            if G.has_edge(neighbors[i], neighbors[j]):
                edges += 1
    cc = 2 * edges / (k * (k-1))
    return cc

cc = {node: clustering_coefficient(G, node) for node in G.nodes()}

n_cc1 = pd.DataFrame(cc.items(),
                     columns=['Noeud', 'Coefficient de clustering'])
print(f"Le coefficient de clustering de chaque noeud sans la fonction predefinie:")
n_cc1
```

```python
# Confirmation que les deux methodes donnent les memes resultats
n_cc.equals(n_cc1)
```

```python
# 4. Coefficient de clustering de graphe G
# Avec la fonction predefinie
G_cc = nx.average_clustering(G)
print(f"Le coefficient de clustering de G est: {G_cc: .4f}")
```

```python
# Coefficient de clustering de graphe G sans la fonction predefinie
def clustering_coefficient_G(G):
    nodes = list(G.nodes())
    sum_clust_coeff = 0
    for node in nodes:
        sum_clust_coeff += clustering_coefficient(G, node)
    return sum_clust_coeff / len(nodes)

G_cc1 = clustering_coefficient_G(G)
print(f"Le coefficient de clustering de G sans la fonction predefinie est: {G_cc1: .4f}")
```

```python
# Degré de centralité de chaque noeud sans la fonction predefinie
def degree_centrality(G, node):
    return len(list(G.neighbors(node))) / (len(G.nodes()) - 1)

deg_cent1 = {node: degree_centrality(G, node) for node in G.nodes()}
deg_cent_n1 = pd.DataFrame(deg_cent1.items(),
                              columns=['Noeud', 'Degré de centralité'])
print(f"Le degré de centralité de chaque noeud sans la fonction predefinie:")
deg_cent_n1
```

```python
# 7. Centralité intermédiaire de chaque noeud
# Avec la fonction predefinie
cent_int = nx.betweenness_centrality(G)
cent_int_n = pd.DataFrame(cent_int.items(),
                           columns=['Noeud', 'Centralité intermédiaire'])
print(f"La centralité intermédiaire de chaque noeud:")
cent_int_n
```

```python
# Centralité intermédiaire de chaque noeud sans la fonction predefinie

def centralite_intermed(G):
    # Initialiser le dictionnaire de centralité d'intermédiarité pour tous les nœuds
    betweenness = {node: 0.0 for node in G.nodes()}

    # Boucle sur chaque nœud en tant que source
    for source in G.nodes():

        stack = []  # Pile pour stocker l'ordre de traitement des noeuds
        predecessors = {node: [] for node in G.nodes()}  # Liste des prédécesseurs
        shortest_paths = {node: 0 for node in G.nodes()}  # Nombre de chemins les plus courts
        shortest_paths[source] = 1
        distances = {node: -1 for node in G.nodes()}  # Distance par rapport à la source
        distances[source] = 0

        # Calculer les chemins les plus courts et les distances
        queue = [source]  # File d'attente
        while queue:
            current = queue.pop(0)
            stack.append(current)
            for neighbor in G.neighbors(current):
                # Si le voisin n'a pas encore été visité
                if distances[neighbor] == -1:
                    queue.append(neighbor)
                    distances[neighbor] = distances[current] + 1
                # Si un chemin le plus court passe par le nœud actuel
                if distances[neighbor] == distances[current] + 1:
                    shortest_paths[neighbor] += shortest_paths[current]
                    predecessors[neighbor].append(current)
        
        # Initialiser les dépendances pour la phase d'accumulation
        dependency = {node: 0.0 for node in G.nodes()}
        
        # Propager les dépendances en utilisant la pile
        while stack:
            node = stack.pop()
            for pred in predecessors[node]:
                dependency[pred] += (shortest_paths[pred] / shortest_paths[node]) * (1 + dependency[node])
            if node != source:
                betweenness[node] += dependency[node]
    
    # Normaliser les valeurs de centralité d'intermédiarité
    scale = 1 / ((len(G) - 1) * (len(G) - 2)) if len(G) > 2 else 1
    for node in betweenness:
        betweenness[node] *= scale
    # Convertir les résultats en DataFrame
    ci = pd.DataFrame(
        {"Noeud": list(betweenness.keys()), "Centralité intermédiaire": list(betweenness.values())}
    )
    
    return ci

ci = centralite_intermed(G)
print(f"La centralité intermédiaire de chaque noeud sans la fonction predefinie:")
ci
```

```python
# Importation des bibliothèques pour la 2eme fois si on veut executer seulement la 2eme partie
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Creation d'un graphe orienté A
A = nx.DiGraph()

# Les sommets
sommets_A = [1, 2, 3, 4, 5, 6, 7, 8]
A.add_nodes_from(sommets_A)

# 1:A / 2:B / 3:C / 4:D / 5:E / 6:F / 7:G / 8:H
# Les aretes
aretes_A = [
    (1,2),(1,4),(1,6),(1,5),
    (2,3),(2,8),
    (3,1),(3,5),(3,8),
    (4,3),(4,5),
    (5,8),(5,7),(5,6),
    (6,4),
    (7,6),(7,8),(7,4),
    (8,7)]

A.add_edges_from(aretes_A)

# Visualisation de graphe A
pos = nx.spring_layout(A, 
                       seed=24)
# seed pour avoir la meme disposition des noeuds

nx.draw(A,
        pos,
        with_labels=True,
        node_color='lightgreen',
        font_weight='bold',
        font_size=10,
        node_size=500)
plt.title("Graphe A")
plt.show()
```

```python
# 2. Matrice d'adjacence de A
matrice_adj_A = nx.adjacency_matrix(A).todense()

nodes_A = list(A.nodes())
df_adj_A = pd.DataFrame(matrice_adj_A, 
                        index=nodes_A,
                        columns=nodes_A)

print(f"Matrice d'adjacence de A:")
df_adj_A
```

```python
# 3. Matrice transposée de A
A_transpose = A.reverse()
matrice_adj_A_transpose = nx.adjacency_matrix(A_transpose).todense()
matrice_adj_A_transpose = pd.DataFrame(matrice_adj_A_transpose, 
                                      index=nodes_A, 
                                      columns=nodes_A)
matrice_adj_A_transpose
```

```python
# 4. a. Algorithme PageRank naîf
adj_A = df_adj_A.copy()
sums = adj_A.sum(axis=0)
adj_A.loc['somme'] = sums
print('Algorithme PageRank naîf:')
adj_A
```

```python
# b. PageRank pondéré
def PageRank_p(df):
    df1 = df.copy()
    sums = df1.sum(axis=1)
    df1['somme'] = sums
    df1 = df1.div(df1['somme'], axis=0)
    df1 = df1.drop(columns=['somme'])
    df1 = df1.round(2)
    sums = df1.sum(axis=0)
    df1.loc['somme'] = sums
    return df1

print('PageRank pondéré:')
PageRank_p(df_adj_A)
```

```python
# 5. Application de algorithme de HITS pour afficher les noeuds par ordre décroissant
# a. de l'autorité
hits = nx.hits(A)
authority = hits[1]

df_authority = pd.DataFrame(authority.items(), columns=['Noeud', 'Autorité'])
df_authority_ord = df_authority.sort_values(by='Autorité', ascending=False)

print(f"Classification des noeuds par ordre décroissant de l'autorité:")
df_authority_ord
```

```python
# b. de l'hub
hub = hits[0]
df_hub = pd.DataFrame(hub.items(), columns=['Noeud', 'Hub'])
df_hub_ord = df_hub.sort_values(by='Hub', ascending=False)

print(f"Classification des noeuds par ordre décroissant de l'hub:")
df_hub_ord
```

<h2>Analyse des résultats</h2>
<h3>1. PageRank naïf</h3>
Le PageRank naïf montre que la page 8 est la plus influente, avec le score le plus élevés 4.<br>
Les noeuds 4, 5 et 6 ont des scores de 3, suivis de 3 et 7 à 2.<br>
1 et 2 ont des scores de 1, indiquant une influence minimale.<br>
<h3>2. PageRank pondéré</h3>
Le PageRank pondéré révèle que 4 et 8 restent dominantes. <br>
Les pages 7 et 3 gagnent en influence grâce à leurs connexions stratégiques, on ne peut pas arriver a 1 sans passer par 3.<br>
<h3>3. HITS : Autorité</h3>
Les pages 8 et 6 ont les meilleures autorités, signifiant qu'elles sont les plus citées par des hubs importants.<br>
1, 3, et 7 ont des scores plus faibles.<br>
<h3>4. HITS : Hub</h3>
La page 1 est le meilleur hub, suivie de 7 et 5.<br>
8 a un faible score de hub, indiquant un faible rôle dans la distribution de liens.<br>
<h3>Comparaison </h3>
PageRank pondéré est plus précis que la méthode naïve en tenant compte de la qualité des liens.<br>
HITS distingue les pages selon leur rôle (autorité vs hub), complétant bien PageRank.<br>
Ensemble, ces deux algorithmes offrent une analyse plus riche du réseau.
